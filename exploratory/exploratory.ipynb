## Libs utilizadas na exploração dos dados

```python
import pandas as pd
from pathlib import Path
```

## Definição do diretório onde está a tabela bruta e onde sera adicionada a tabela transformada 

```python
BASE_DIR = Path().resolve().parent
INPUT_PATH = BASE_DIR / "data" / "output" / "precos_combustivel_bruto.parquet"
OUTPUT_PATH = BASE_DIR / "data" / "output" / "precos_combustivel_transformado.parquet"
```

## Define os replaces dos nomes dos campos para melhor legibilidade
Antigo_Nome : Novo_Nome

```python
NEW_NAME_OF_COLUMNS = {
    "Regiao - Sigla": "Regiao",
    "Estado - Sigla": "Estado",
    "Municipio": "Cidade",
    "Revenda": "Revendedor",
    "CNPJ da Revenda": "CNPJ do Revendedor",
    "Produto": "Tipo de Combustível",
    "Bandeira": "Fornecedor"
}
```

## Leitura do arquivo e informações iniciais para base da análise

```python
df = pd.read_parquet(INPUT_PATH)
df.info()
df.head(5)
```

## Limpeza básica nos dados ajustando nome das colunas, removendo uma coluna desnecessária e adicionando tipagem as dados que forem necessários.

```python
df.rename(columns=NEW_NAME_OF_COLUMNS, inplace=True)
df.drop('Valor de Compra', axis=1, inplace=True)
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',', '.').astype('float')
df['Data da Coleta'] = pd.to_datetime(df['Data da Coleta'], format='%d/%m/%Y', errors='coerce')
df = df[df['Data da Coleta'] != 'NaT']
df.head(5)
```

## Identificando quantidade de nulos para limpeza identificando onde pode ser nulo e onde não pode ser nulo

```python
df.isna().sum()
```

```python
df.dropna(subset=['Tipo de Combustível', 'Revendedor', 'CNPJ do Revendedor', 'Fornecedor', 'Valor de Venda', 'Data da Coleta'], inplace=True)
df.isna().sum()
```
